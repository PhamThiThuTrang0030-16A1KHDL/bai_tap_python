```python
import numpy as np
# 1
with open('positions.txt', 'r', encoding = 'utf-8-sig') as file:
    positions = file.read()
    list_positions = positions.split(',')
    np_positions = np.array(list_positions)
with open('heights.txt', 'r', encoding = 'utf-8-sig') as f:
    heights = f.read()
    np_heights = np.fromstring(heights,sep=',')

print(f"Kiểu dữ liệu của np_positions: {np_positions.dtype}")
print(f"Kiểu dữ liệu của np_heights: {np_heights.dtype}")

```

```python
# 2
gk_heights = np_heights[np_positions == 'GK']
print(gk_heights)
print(f"Chiều cao trung bình của các GK: {np.mean(gk_heights)}")
```

```python
# 3
other_heights = np_heights[np_positions != 'GK']
print(f'chiều cao trung bình của những vị trí khác: {np.mean(other_heights)}')
```

```python
# 4
players = np.zeros(len(np_positions), dtype={'names':('position', 'height'), 'formats':('U5', 'f8')})
players['position'] = np_positions
players['height'] = np_heights
print(players)
```

```python
# 5
players = np.sort(players, order='height')
print(f"Vị trí có chiều cao thấp nhất: {players[0]['position']}, Chiều cao: {players[0]['height']}")
print(f"Vị trí có chiều cao cao nhất: {players[-1]['position']}, Chiều cao: {players[-1]['height']}")
```
